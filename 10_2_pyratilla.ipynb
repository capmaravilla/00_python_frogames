```python
coins =450000

def convert_to_coins(cantidad, divisa):
    if divisa == "moneda":
        return cantidad * 1
    if divisa == "coralina":
        return cantidad * 0.75 
    if divisa == "rupia":
        return cantidad * 0.55 
    if divisa == "chuche":
        return cantidad * 1.30 
    if divisa == "lita":
        return cantidad * 2.15 
    if divisa == "gema":
        return cantidad * 1.70 
    if divisa == "corona":
        return cantidad * 0.95 

coralinas = convert_to_coins(1310, "coralina") 
rupias = convert_to_coins(2025, "rupia") 
chuches = convert_to_coins(1520, "chuche") 
litas = convert_to_coins(970, "lita")
gemas = convert_to_coins(1975, "gema")
coronas = convert_to_coins(1760, "corona")

coins += 1050 + coralinas + rupias + chuches + litas + gemas + coronas 

def currency_converter(cantidad, div_inicial, div_final):
    to_moneda = convert_to_coins(cantidad, div_inicial)
    if div_final == "moneda":
        return to_moneda / 1
    if div_final == "coralina":
        return to_moneda / 0.75 
    if div_final == "rupia":
        return to_moneda / 0.55 
    if div_final == "chuche":
        return to_moneda / 1.30 
    if div_final == "lita":
        return to_moneda / 2.15 
    if div_final == "gema":
        return to_moneda / 1.70 
    if div_final == "corona":
        return to_moneda / 0.95 

islands = {"Isla Alegre", "Isla Golosa", "Isla Espesura", "Isla Arrecife", "Isla Lejana", "Isla Torva", "Isla Verde"}
visited_islands = {"Isla Alegre", "Isla Arrecife", "Isla Espesura"}
unvisited_islands = islands - visited_islands

def num_even_odd_words(texto):
    sample = texto.split(' ')
    result = {
        'even' : 0,
        'odd' : 0
    }
    for i in sample:
        if len(i) % 2 == 0:
            result["even"] += 1
        else:
            result["odd"] += 1
    return result

num_even_odd_words('Hola caracola donde dices que me ves')
```
