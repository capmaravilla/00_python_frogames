```python
import pandas as pd

words = ["pulpo", "zaraz", "nadan", "amor", "asa"]
data = {
    "word": words,
    "length": map(len, words),
    "start": map(lambda w: w[0], words),
    "end": map(lambda w: w[-1], words),
    "isPalindrome": map(lambda x: x == x[::-1], words),
}

words = pd.DataFrame(data)
print(words)

words2 = words.set_index("word")
print(words2)
words2.index.names = [None]
print(words2)
```

```python
filas, columnas = words2.shape
for i in range(columnas):
    print(f"==== {list(words2)[i].upper()} ====\n{words2.iloc[:, i]}")

"""
# Otra forma de hacer la primera parte del ejercicio
for col in list(df):
    print(f"---{col.upper()}---")
    print(df[col], end="\n")
"""


word3 = words2.iloc[:, 1:]
word3
```
